```python
def loadData(name : str = "data/fullCatalog.csv"):
    return pd.read_csv(name)


full = loadData("data/fullCatalog.csv")
slim = loadData("data/slimCatalog.csv")
```

```python
genres = ["electr", "pop", "rap", "hip", "deep", "r&b", "soul", "reggae", "latin", "country", "funk", "jazz", "classical","folk", "blues", "rock", "metal", "garage", "punk", "indie", "techno", "trance", "dance", "house", "disco", "mellow gold", "christ", "grunge"]
```

```python
slim.head()
```

```python
full.head()
```

```python
genresOnly = slim[["id", "genreArr"]]
```

```python
genresOnly.head()
```

```python
master = full.merge(genresOnly, how="inner", on="id")
```

```python
genres[1]
```

```python
master.columns
```

```python
import ast
new = master.drop(["Unnamed: 0", "Unnamed: 0_x", "Unnamed: 0_y"], axis=1)
```

```python
def genreArrToString(genreArr = "[1, 2]"):
    return str([genres[i] for i in ast.literal_eval(genreArr)])

new["genreArrStr"] = new.genreArr.map(genreArrToString)
```

```python
new = new.drop("genreArr", axis=1)
new.head()
```

```python
new.columns
```

```python
for c in ['acousticness', 'danceability','energy', 
          'instrumentalness', 'liveness', 
          'mode', 'speechiness', 'valence']:
    new[c] = (new[c] * 1000).astype(int)
```

```python
new["key"] = new["key"].astype(int)
new["time_signature"] = new["time_signature"].astype(int)
new.head()
```

```python
new.to_csv("data/new.csv", index=False)
```

```python
new.drop(["album_id", "album_name"], axis=1).to_csv("data/newNoAlbum.csv")
```
